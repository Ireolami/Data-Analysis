```python
import pandas as pd
melb = pd.read_csv("Melbourne_housing_FULL.csv")
mel_price = melb.Price.fillna(melb.Price.mean())
melb_room = pd.DataFrame(melb.Rooms, columns=["Rooms"])
melb_bed = melb.Bedroom2.fillna(melb.Bedroom2.mean())
melb.corr()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(melb_room, melb_bed, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
# y_pred.shape
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
intercept = lr.intercept_
print(r2)
print(mse)
print(mae)
print(intercept)
```

```python
# How the landsize, buildingarea, newsuburb and distance affect the price

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
melb['newsuburb'] = le.fit_transform(melb["Suburb"])
melb.BuildingArea = melb.BuildingArea.fillna(melb.BuildingArea.mean())
melb.Landsize = melb.Landsize.fillna(melb.Landsize.mean())
melb.Distance = melb.Distance.fillna(melb.Distance.mean())

allfeature = melb[["Landsize", "BuildingArea", "newsuburb", "Distance"]]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(allfeature, mel_price, test_size = 0.2)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
# y_pred.shape
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
intercept = lr.intercept_
print(r2)
print(mse)
print(mae)
print(intercept)
```

```python
# Working with titanic dataset.
import pandas as pd
titanic = pd.read_csv("Titanic-Dataset.csv")
titanic["Embarked"] = titanic["Embarked"].fillna(method="ffill")
titanic
```

```python
from sklearn.preprocessing import LabelEncoder
#Label encoder is purposely used for transformation of text to number
le = LabelEncoder()
titanic['newEmb'] = le.fit_transform(titanic["Embarked"])
titanic['newSex'] = le.fit_transform(titanic["Sex"])

titanic
```

```python
feature = titanic[["Pclass", "SibSp", "newSex", "Parch", "newEmb"]]
target = titanic["Survived"]

from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(feature, target, random_state=1, train_size=0.8)

Xtrain.shape
```

```python
from sklearn.naive_bayes import GaussianNB # 1. choose model class
model = GaussianNB() # 2. instantiate model
model.fit(Xtrain, ytrain) # 3. fit model to data
y_model = model.predict(Xtest) # 4. predict on new data

# print("Predicted \n", y_model)
# print("Original label \n", ytest.values)
Xtest
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ytest, y_model)
```

```python
newfeature = pd.DataFrame([[2, 1, 1, 0, 2]], columns=["Pclass", "SibSp", "newSex", "Parch", "newEmb"])
y_model = model.predict(newfeature)
y_model
```
