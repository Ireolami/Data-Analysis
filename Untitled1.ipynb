```python
import yfinance as yf
import pandas as pd
```

```python
#Getting the period of the coin
btc_ticker = "BTC-USD" 
start_date = "2011-01-01" 
end_date = "2024-12-05"  

#Historical data downloading
btc_data = yf.download(tickers=btc_ticker, start=start_date, end=end_date, interval="1d")

# Saving the data
csv_filename = "btc_historical_data.csv"
btc_data.to_csv(csv_filename)

```
